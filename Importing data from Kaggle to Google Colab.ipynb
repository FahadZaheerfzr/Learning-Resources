# Downloading Data from Kaggle directly on Google Colab

Google Colab offers free GPU but feeding data into the colab can be a tedious process requiring you to manually upload on colab or drive. But we can directly import the data from kaggle on Google Colab. All we need to do is create an account on Kaggle. Go to My Profile and create an API Token for your profile. Make sure you have a verified account with verified phone number and email address.

Kaggle API.png



The snippet above shows the section in my profile where you can generate API Token for yourself.

After generating the API Token copy the content of the API Token and Run the following command with your username and your key in the placeholder

```
!mkdir /root/.kaggle
!echo '{"username":"Your_Username","key":"Your_Key"}' > /root/.kaggle/kaggle.json
!pip install kaggle
!kaggle datasets download -d kmader/skin-cancer-mnist-ham10000
```

```
!unzip -q /content/skin-cancer-mnist-ham10000.zip -d dataset
```

**Note**
The Command in line 6 in the above shell is different for different competitions and datasets. This command can be generated from copy API Command option of each dataset or competition. The format of this command is *!API Command*
